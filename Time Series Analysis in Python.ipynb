[Google Trends](https://trends.google.com/trends/) allows users to see how often a term is searched for. We downloaded a file from Google Trends containing the frequency over time for the search word "diet".

```python
diet = pd.read_csv('datasets/diet.csv',parse_dates=['Date'], index_col='Date')
diet
```

```python
diet.info()
```

A first step when analyzing a time series is to visualize the data with a plot. 

```python
diet.plot()
plt.title('The entire time series diet')
plt.show()
```

There is a gradual decrease in searches for "diet" throughout the calendar year, hitting a low around the December holidays, followed by a spike in searches around the new year as people make New Year's resolutions to lose weight.

We will explore for the year 2012

```python
diet[diet.index.year==2012].plot(title="2012 Time series", legend=False)
plt.show()
```

searches for 'diet' spiked up after the holidays every year.
